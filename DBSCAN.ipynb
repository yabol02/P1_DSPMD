Práctica 1: Algoritmo DBSCAN
El siguiente código crea dos nubes de puntos con forma de media luna.
A continuación aplica el algoritmo DBSCAN con una distancia entre clústeres de 0.3 y un número mínimo de muestras para formar un cluster de 5.

```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Genera datos de ejemplo (en forma de "luna")
X, _ = make_moons(n_samples=200, noise=0.05, random_state=0)

# Crea un objeto DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)

# Aplica el algoritmo
y_dbscan = dbscan.fit_predict(X)

# Visualiza los resultados
axes[0].scatter(X[:, 0], X[:, 1], c='r')
axes[0].set_title("Original data")
axes[1].scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis')
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")
axes[1].set_title("DBSCAN Clustering");
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

noise = 0.1
X, _ = make_moons(n_samples=200, noise=noise, random_state=0)
dbscan = DBSCAN(eps=0.18, min_samples=2)
y_dbscan = dbscan.fit_predict(X)

fig.suptitle(f"DBSCAN Clustering with Noise={noise}", fontsize=16)
axes[0].scatter(X[:, 0], X[:, 1], c='r')
axes[0].set_title("Original data")
axes[1].scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis')
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")
axes[1].set_title("DBSCAN Clustering");
```
